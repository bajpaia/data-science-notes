## Types of Data
- Categorical
- Numerical
    - Discrete
    - Continous
### Types of Data measurement
- Nominal: Data is categorical with no ranking (e.g gender 0- male 1- female) no arthmetic operations can be applied ot these, these can only be classified and counted (non-parametric statistical tests only)

- Ordinal: Distinct Categories with ranks e.g. level of staisfication (1-5)no arthmetic operations can be applied on these variables, these can be classified counted and ranked (non-parametric statistical tests only)

- interval - ordered without zero point (e.g. Years, farenheit tempertature) addition and substraction can be applied, these can be classified counted, ranked, added and substracted (parametric statistical tests)

- ratio - ordered with zero point (e.g weight, age, salary) we can do a lot of arthemtic operation depending on what would give the data meaning,  these can be classified counted, ranked, added ,substracted, multiplied and divided(parametric statistical tests)

## Central Tendency and Dispersion

Measures of central tendency yield information about particular location in a group of numbers (using a single number to define a set of data e.g mean, median, weighted mean and percentile) (Dispersion- skewness, Kurtosis, Range, interquartile range, variance, standard score, coefficient of variation)

### Central Tendency

#### Arithmetic mean

Average of the data in a group, cannot be used for nominal and ordinal data, sensitive to outliers. (Summation of all values/number of values)

#### Sample mean

Average of a subset of the whole data

#### Group mean

Weighted average of midpoints of the group (e.g. 20-30 the midpoint would be 25), Summation of frequency*midpoints/ number of data points 

$\frac{ f_i*m_i}{\sum_{i=1}^{n} f_i}$

#### Weighted average

To compute a weighted average you need to multiply the weight of the corresponding datapoint with the datapoint and divide by the sum of weights

$\frac{\sum_{i=1}^{n} w_i*x_i}{\sum_{i=1}^{n} w_i}$

#### Median

It is a central value of a dataset that is unaffected by outliers

e.g for a sorted array of odd numbers the median is the middle value$\frac{n+1}{2}$

For a sorted array of even numbers the median is average of the middle two terms.

It is generally used to find the middle point in a skewed dataset

#### Mode

It is the most frequently appearing value in a dataset.


#### Percentiles

Divide the data into 100 parts.

>e.g 90th percentile would mean that 90% of the data in a sorted array is under the value x where $x = X[len(X) * 0.90]$

The median and 50th percentile have the same value

>If x is the median value $x =X[len(X)*0.50]$

if the percentiles location ($len(x) * Percentile/100 $) is not a whole number we will use 
> $x = X[len(x) * Percentile/100 +1]$

### Dispersion

Dispersion measures are values that measure the variability of the data, These give an idea about the reliability of the central tendency values.

#### Range

The difference between the largest and smallest values, ignores all other values between the max and min values

$max(X) - min(X)$

where X is an array of the Data


#### Quartiles

Divide the data into 4 groups:
- Q<sub>1</sub>: **25%** of the data lies in the ***first quartile*** = **25<sup>th</sup> percentile**
- Q<sub>2</sub>: **50%** of the data lies in the ***second quartile*** = **50<sup>th</sup> percentile**
- Q<sub>3</sub>: **75%** of the data lies in the ***third quartile*** = **75<sup>th</sup> percentile**
- Q<sub>4</sub>: **100%** of the data lies in the ***fourth quartile*** = **100<sup>th</sup> percentile**

>#### Inter-quartile range
$IQR = Q_3 - Q_1$ (This value are not affected by outliers)


#### Mean Absolute Deviation from mean

The difference of each data value from the mean

$ \frac {\sum_{i=1}^{n}|x_i - µ|}{n} $


#### Variance

The average of squared deviations from the mean is called variance, the squaring of the values helps by penalising the values which are further away from the mean.

$\sigma^{2} = \frac{\sum_{i=1}^{n} (x_i - µ)^2}{N}$

#### Standard Deviation

Std deviation helps compare variance values. 
>Example use cases:
- Risk analysis, higher std dev means higher risk
- Quality control lower std dev means higher quality
- Process capatbility, lower std deviation means higher process capability.

$ std = \sqrt{\sigma^{2}}$


### Interpretting Std Deviation

####  The Emperical Rule

>For a **Normal Distribution**
- **68%** of all data will be within **$1\sigma$**
- **95%** of all data will be within **$2\sigma$**
- **99.7%** of all data will be within **$3\sigma$**



>- $µ ± 1*\sigma ≈ 68%$ of all data
- $µ ± 2*\sigma ≈ 95% of Data$ of all data
- $µ ± 3*\sigma ≈ 99.7% of Data$ of all data

where µ is mean and $\sigma$ is standard deviation


#### Chebyshev's Theorem
(not often used because interval is very wide)

For all Distributions:
$1-\frac{1}{k^2}$ for $k>1$ 

Here 
$x = (1-\frac{1}{k^2})*100$ 

Therefore, **atleast x per-cent  of the data is in $k$ standard deviations** 


#### Coefficient of Variation

$C.V = \frac{\sigma}{µ}(100)$

where $\sigma$ is std deviation and µ is the mean.

This is the measure of relative dispersion, this can be used to compare two distributions. Lower the CV, the closer the distribution is to a normal distribution
 > In risk analysis **lower Coef of variation** is **lower risk**


### Measures of Shape

#### Skewness
- Absence of symmetry (**Symmetry** occurs when **Mean = Median = Mode**)
- Extreme values on one side of the distribution
    - Right Skew(positive skew): Mode<Median<Mean
    - Left Skew(negative skew): Mean<Median<Mode 

> Note:
    $S = \frac{3(µ-M_d)}{\sigma}$<br></br>
> Where:
- S is coefficient of skewness
- µ is the mean 
- $M_d$ is the median
- $\sigma$ is the standard deviation
- S < 0: Left skew
- S > 0: Right skew
- S = 0: Symmetric 
#### Kurtosis
- Leptokurtic: tall and thin distribution
- Mesokurtic: Normal distribution(bell shaped)
- Platykurtic: flatbread  and spread out
